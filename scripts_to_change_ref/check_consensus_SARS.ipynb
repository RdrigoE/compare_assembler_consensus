```python
with open("validated_variants.csv", 'r', encoding='utf-8') as handler:
    dic_info: dict[int,list[str]] = {}
    for line in handler.readlines():
        name,_, pos,_1, ref, alt, *remaining = line.split(',')
        if name == "demo_SARSCoV2_001":
            dic_info[int(pos)] = [ref,alt]
    print(len(dic_info.keys()))
```

```python
from Bio import SeqIO
from Bio.Seq import MutableSeq
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
with open("new_reference.fasta", 'r', encoding='utf-8') as handler:
    ref = list(SeqIO.parse(handler, "fasta"))[0]
    mut = MutableSeq(ref.seq)
    for key, v in dic_info.items():
        print(f"Pre:  {mut[key-1]}")
        print(f"ref=  {v[0]} \t alt={v[1]}")
        mut[key-1] = v[1]
        print(f"Pos:  {mut[key-1]}")
        print("-- " * 20)
    MUTATED_SEQ = SeqRecord(
        Seq(str(mut)),
        id = ref.id,
    )
    print(MUTATED_SEQ)
with open("new_reference.fasta", 'w') as output_handler:
    SeqIO.write(MUTATED_SEQ, output_handler, "fasta")
```

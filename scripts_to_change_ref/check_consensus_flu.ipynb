```python
with open("validated_variants.csv", 'r', encoding='utf-8') as handler:
    dic_info: dict[int,list[str]] = {}
    for line in handler.readlines():
        name,id, pos,_1, ref, alt, *remaining = line.split(',')
        if name == "demo_influenza_01":
            dic_info[int(pos)] = [id,ref,alt]
```

```python
from Bio import SeqIO
from Bio.Seq import MutableSeq
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
with open("user/references/A_H3N2_A_Victoria_361_2011.fasta", 'r', encoding='utf-8') as handler:
    ref = list(SeqIO.parse(handler, "fasta"))
    whole_genome = []
    for entry in ref:
        print(entry.id)
        mut = MutableSeq(entry.seq)
        for key, v in dic_info.items():
            
            if v[0] == entry.id:
                print(f"Pre:  {mut[key-1]}")
                print(f"ref=  {v[1]} \t alt={v[2]}")
                mut[key-1] = v[2]
                print(f"Pos:  {mut[key-1]}")
                print("-- " * 20)
        mutated_entry = SeqRecord(
            Seq(str(mut)),
            id = entry.id,
        )
        whole_genome.append(mutated_entry)

with open("new_reference.fasta", 'w') as output_handler:
    SeqIO.write(whole_genome, output_handler, "fasta")
```
